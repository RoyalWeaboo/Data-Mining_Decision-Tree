pertama mengimport library numpy, pandas dan scikit-learn

```python
import numpy as np
import pandas as pd
from sklearn import tree
```

selanjutnya membaca dataset menggunakan pandas

```python
irisDataset = pd.read_csv('Dataset_Iris_Modified.csv', delimiter=';', header=0)
print(irisDataset)
```

kemudian ubah kelas Species menjadi unique integer

```python
irisDataset["Species"] = pd.factorize(irisDataset.Species)[0]
print(irisDataset)
```

kemudian hapus kolom id

```python
irisDataset = irisDataset.drop(labels="Id", axis=1)
print(irisDataset)
```

mengubah dataframe ke array numpy agar memudahkan untuk komputasi

```python
irisDataset = irisDataset.to_numpy()
print(irisDataset)
```

kemudian membagi dataset 80% untuk training dan 20% untuk testing

```python
dataTraining = np.concatenate((irisDataset[0:80, :], irisDataset[100:180, :]), axis=0)
dataTesting = np.concatenate((irisDataset[80:100, :], irisDataset[180:200, :]), axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

kemudian lakukan pemecahan dataset ke input dan label

```python
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]
```

Selanjutnya mendefinisikan Decision Tree Classifier menggunakan function DecisionTreeClassifier()

```python
model = tree.DecisionTreeClassifier()
```

Kemudian dilanjutkan mentraining model untuk inputTraining terhadap labelTraining

```python
model = model.fit(inputTraining, labelTraining)
```

kemudian dilanjutkan mempredikisi input data testing

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

kemudian hitung akurasinya

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, "data")
print("Prediksi Salah : ", prediksiSalah, "data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)*100, "%")
```
